```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# save driver count from city data
# driver_counts = city_data[['city', 'driver_count']]
# driver_counts = city_data['driver_count']
grp = city_data.groupby('type')
tot_driversXcity_type = grp['driver_count'].sum()
```

```python
# Combine the data into a single dataset
merged_data = pd.merge(ride_data, city_data, how="left", on=["city", "city"])

# Display the data table for preview
merged_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types (Urban, Suburban, Rural)
merged_data['type'].unique()

# separate DFs by TYPE
urban_data = merged_data.loc[merged_data['type']=='Urban', : ]
suburb_data = merged_data.loc[merged_data['type']=='Suburban', : ]
rural_data = merged_data.loc[merged_data['type']=='Rural', : ]

# groupby city to get avg fares and ride count
urban_grp_cities = urban_data.groupby('city')
urban_fares = urban_grp_cities['fare'].mean()
urban_ride_count = urban_grp_cities['ride_id'].count()
urban_driver_count = urban_grp_cities['driver_count'].mean()*10

suburban_grp_cities = suburb_data.groupby('city')
suburban_fares = suburban_grp_cities['fare'].mean()
suburban_ride_count = suburban_grp_cities['ride_id'].count()
suburban_driver_count = suburban_grp_cities['driver_count'].mean()*10

rural_grp_cities = rural_data.groupby('city')
rural_fares = rural_grp_cities['fare'].mean()
rural_ride_count = rural_grp_cities['ride_id'].count()
rural_driver_count = rural_grp_cities['driver_count'].mean()*10
```

```python

# Build the scatter plots for each city types

# urban scatter plot
# s=  sets The size of each point on our plot is determined by their x value
plt.scatter(urban_ride_count, urban_fares, marker="o", facecolors="coral", edgecolors="black", alpha=0.5, s=urban_driver_count)
plt.scatter(suburban_ride_count, suburban_fares, marker="o", facecolors="lightskyblue", edgecolors="black", alpha=0.5, s=suburban_driver_count)
plt.scatter(rural_ride_count, rural_fares, marker="o", facecolors="gold", edgecolors="black", alpha=0.5, s=rural_driver_count)


# plt.ylim(0, 1)
# plt.xlim(0, x_limit)


# Incorporate the other graph properties
plt.xlabel("Number of Rides per City")
plt.ylabel("Average Fare per Ride")
plt.title('Pyber Ride Sharing Data (2016)')
plt.grid(b=True, which='major', color='grey', linestyle='-', alpha=.25)

# Create a legend
lgnd = plt.legend(['Urban','Suburban','Rural'])
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]


# Incorporate a text label regarding circle size
# need to work on this
plt.text(42,30,s='''Note:
Circle size correlates with driver count per city''')

# Save Figure

plt.savefig("2016dataplot.png")
plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
grouped_df = merged_data.groupby('type')
tot_fares = grouped_df['fare'].sum()


# extracting type labels
types_list = []
types = grouped_df['type'].unique()
for n in types:
    types_list.append(types[n][0][0])
    
# Build Pie Chart
colors = ["gold","lightskyblue","lightcoral"]
explode = (0,0,0.1)
plt.pie(tot_fares, explode=explode, labels=types_list, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Fares by City Type')

# Save Figure
plt.savefig("2016_percent_fares_city_type.png")
plt.show()
```

```python
# example Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
tot_rides = grouped_df['ride_id'].count()

# Build Pie Chart
plt.pie(tot_rides, explode=explode, labels=types_list, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Rides by City Type')


# Save Figure
plt.savefig("2016_percent_rides_city_type.png")
plt.show()
```

```python
# Show Figure
plt.show()
```
